```python
from nbodykit.lab import UniformCatalog, FFTPower
#calculate fluxes, do things in flux space
# pm_red = nbodykit.source.mesh.array.ArrayMesh(flux_red, 320)
pm_ex = nbodykit.source.mesh.array.ArrayMesh(flux_ex, 320)

```

```python
m_array = np.load("pkell_red.npy")
l0 = np.where(m_array[:,1]==0)[0]
k_l0 = m_array[l0,0]
Pk_l0 = m_array[l0,2]
l2 = np.where(m_array[:,1]==2)[0]
k_l2 = m_array[l2,0]
Pk_l2 = m_array[l2,2]
l4 = np.where(m_array[:,1]==4)[0]
k_l4 = m_array[l4,0]
Pk_l4 = m_array[l4,2]
```

```python
# r_red_2d = FFTPower(pm_red, mode='2d', dk=0.1, kmin=0.0001, Nmu=4, los=[0,0,1])
# Pkmu_red = r_red_2d.power

r_ex_2d = FFTPower(pm_ex, mode='2d', dk=0.1, kmin=0.01, Nmu=5, los=[0,0,1], poles=[0,2,4])
poles = r_ex_2d.poles
plt.figure(figsize=(8,8))

# plot each mu bin
lines = ["r","g","b","k","y"]
# for i in range(Pkmu_red.shape[1]):
#     Pk = Pkmu_red[:,i] # select the ith mu bin
#     label = r'$\mu$=%.1f' % (Pkmu_red.coords['mu'][i])
#     plt.loglog(Pk['k'], Pk['power'].real,lines[i], label=label)
plt.plot(k_l0, k_l0 * Pk_l0, linestyle='dashed')
# plt.plot(k_l2, -k_l2 * Pk_l2, linestyle='dashed')

# format the axes
plt.legend(loc=0, ncol=2)
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P(k, l)$ [$h^{-3}\mathrm{Mpc}^3$]")

#plot realspace as little dots on top
for ell in [0]:
    label = r'$\ell=%d$' % (ell)
    P = poles['power_%d' %ell].real
    # if ell == 0: P = P - poles.attrs['shotnoise']
    plt.plot(poles['k'], poles['k'] * P, label=label)

plt.yscale("log")
```
